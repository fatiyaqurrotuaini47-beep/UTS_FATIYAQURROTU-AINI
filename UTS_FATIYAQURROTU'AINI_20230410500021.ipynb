```python
# Import & Download Resource

import pandas as pd
import re
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# LOAD DATASET

df = pd.read_csv('/content/Ganjil.csv', sep=';')

# bersihkan nama kolom
df.columns = df.columns.str.strip()

# hapus data kosong
df = df.dropna(subset=[df.columns[1]])

df.head()
```

```python
# =========================
# 1. TEXT CLEANING
# =========================
def text_cleaning(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['cleaning'] = df.iloc[:,1].apply(text_cleaning)

print("===== 1. TEXT CLEANING ====")
# Fix: Iterate over the values of the 'cleaning' column, not by potentially missing index labels
for i, cleaned_text in enumerate(df['cleaning'].head(5)):
    print(f"Data {i+1}:", cleaned_text)
```

```python

# 2. TOKENIZATION

from nltk.tokenize import word_tokenize

df['token'] = df['cleaning'].apply(word_tokenize)

print("\n===== 2. TOKENIZATION ====")
# Fix: Iterate over the values of the 'token' column, not by potentially missing index labels
for i, token_list in enumerate(df['token'].head(5)):
    print(f"Data {i+1}:", token_list)
```

```python
# 3. NORMALIZATION

normal_dict = {
    "gk": "tidak",
    "ga": "tidak",
    "yg": "yang",
    "dgn": "dengan",
    "tdk": "tidak"
}

def normalize(tokens):
    return [normal_dict.get(word, word) for word in tokens]

df['normalized'] = df['token'].apply(normalize)

print("\n===== 3. NORMALIZATION ====")
# Fix: Iterate over the values of the 'normalized' column, not by potentially missing index labels
for i, normalized_list in enumerate(df['normalized'].head(5)):
    print(f"Data {i+1}:", normalized_list)
```

```python
# 4 CASE FOLDING

# fungsi case folding
def case_folding(text):
    text = str(text)      # ubah ke string
    text = text.lower()   # ubah jadi huruf kecil
    return text

# terapkan ke dataset (kolom Berita)
df['case_folding'] = df.iloc[:,1].apply(case_folding)

# tampilkan hasil
# Fix: Iterate over the values of the 'case_folding' column, not by potentially missing index labels
for i, case_folded_text in enumerate(df['case_folding'].head(5)):
    print(f"Data {i+1}:", case_folded_text)
```

```python
# 5. STOPWORD REMOVAL

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stopwords]

df['stopword'] = df['normalized'].apply(remove_stopwords)

print("\n===== 5. STOPWORD REMOVAL ====")
# Fix: Iterate over the values of the 'stopword' column, not by potentially missing index labels
for i, stopword_list in enumerate(df['stopword'].head(5)):
    print(f"Data {i+1}:", stopword_list)
```

```python
# 6. STEMMING (PORTER)
from nltk.stem import PorterStemmer

ps = PorterStemmer()

def stemming(tokens):
    return [ps.stem(k) for k in tokens]

df['stemming'] = df['stopword'].apply(stemming)

print("\n===== 6. STEMMING ====")
# Fix: Iterate over the values of the 'stemming' column, not by potentially missing index labels
for i, stemmed_list in enumerate(df['stemming'].head(5)):
    print(f"Data {i+1}:", stemmed_list)
```

```python
# 7. LEMMATIZATION
# =========================
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'
def lemmatization(tokens):
    tagged_tokens = pos_tag(tokens)
    hasil = []

    for word, tag in tagged_tokens:
        if word.lower() in ['are', 'is', 'an']:
            hasil.append(word.lower())
        else:
            hasil.append(
                lemmatizer.lemmatize(word, get_wordnet_pos(tag))
            )

    return hasil

df['lemmatization'] = df['stemming'].apply(lemmatization)

print("\n===== 7. LEMMATIZATION ====")
# Fix: Iterate over the values of the 'lemmatization' column, not by potentially missing index labels
for i, lemmatized_list in enumerate(df['lemmatization'].head(5)):
    print(f"Data {i+1}:", lemmatized_list)
```
